```python
#importing the image module from matlplotlib library

import matplotlib.image as mpimg

#importing the pyplot module from matlplotlib library

import matplotlib.pyplot as plt
import numpy as np
```

```python
# loading an image an image through matplotlib.image module

img = mpimg.imread('/content/dog.jpeg')
```

```python
type(img)
```

```python
print(img.shape)
```

```python
print(img.dtype)
```

```python
img = (img * 255).astype(np.uint8)

print(img.dtype)
print(img[0,0])
```

```python
print(img)
```

```python
# displaying the image from numpy array

img_plot = plt.imshow(img)
plt.show()
```

Resizing the image using pillow library

```python
from PIL import Image

img = Image.open('/content/dog.jpeg')
img_resized = img.resize((200,200))
```

```python
img_resized.save('dog_image_resized.jpg')
```

```python
# displaying the image from numpy array
img_res = mpimg.imread('/content/dog_image_resized.jpg')
img_res_plot = plt.imshow(img_res)
plt.show()
```

```python
print(img_res.shape)
```

Converting RGB image into GrayScale image using openCV

```python

# importing openCV library here cv stands for computer vision
import cv2
```

```python
img = cv2.imread('/content/dog.jpeg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```

cv2.imshow() will display the image but this is not allowed in google colab
Alternative is **from google.colab.patches import cv2_imshow**

```python
from google.colab.patches import cv2_imshow
```

```python
# displaying the grayscale image

cv2_imshow(grayscale_image)
```

```python
# saving the grayscale image

cv2.imwrite('dog_grayscale_image.jpg',grayscale_image)
```

```python
img_gray = mpimg.imread('/content/dog_grayscale_image.jpg')
img_res_plot = plt.imshow(img_gray)
plt.show()
```
